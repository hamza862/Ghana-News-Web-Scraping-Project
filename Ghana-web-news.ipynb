## Ghana News Web Scrapping Project

```python
## Lets the Import the Required Libraries

import requests
from bs4 import BeautifulSoup as bs
```

```python
## Let's load the webpage

headers = {
    'Accept-Encoding': 'gzip, deflate, sdch',
    'Accept-Language': 'en-US,en;q=0.8',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
}


r = requests.get('https://www.ghanaweb.com/GhanaHomePage/politics/', headers=headers)
```

```python
## Create a beautifu Soup object

soup = bs(r.content)

## Let's Print out the webpage
print(soup.prettify())
```

```python
## Getting the only related (First Page Politics News) html data

news_box = soup.find(class_='afcon-left')

print(news_box.prettify())
```

```python
## Let's extract all the news links from the page

news_links = news_box.select('a')

all_news_links = []
for index, link in enumerate(news_links):
    all_news_links.append(news_links[index]['href'])
```

```python
## Let's print out all the links
all_news_links
```

```python
## Checking the length of the links
len(all_news_links)
```

```python
## Creating the Full path for all the news on the page

Base_path = 'https://www.ghanaweb.com/'
Full_Path = []
for index, link in enumerate(all_news_links):
    Full_Path.append(Base_path + all_news_links[index])
```

```python
## Let's print out the Full Path
Full_Path
```

#### Extracting the Titles of all the links from the Politics Home Page

```python
## Extracting the title of every news link of the page

Title = []

for index, link in enumerate(Full_Path):
    url = link
    response = requests.get(url, headers=headers)
    webpage = bs(response.content)
    Title.append(webpage.find('h1').get_text())
```

```python
## Let's print out all the titles
Title
```

```python
## Checking the length of the titles
len(Title)
```

#### Extract the content of all the links from Politics Home Page

```python
## Extracting all the content of all the news links

Content = []
Final_Content = []

for link in Full_Path:
    url = link
    response = requests.get(url, headers=headers)
    webpage = bs(response.content)
    Text = webpage.find_all('p', attrs={'id': 'article-123'})
    for index, text in enumerate(Text):
        Content.append(Text[index].get_text())
        
for index, content in enumerate(Content):
    Final_Content.append(content.replace('\n', '').replace('\r', ''))
```

```python
## Let's Print out the Final Content
Final_Content
```

```python
## Checking the lenght of the Final Content
len(Final_Content)
```

```python
## Let's dump all the titles and content in one Pandas DataFrame
import pandas as pd

data_news = pd.DataFrame(list(zip(Final_Title, Final_Content)),
               columns =['Title', 'Content'])
```

#### Extracting the title and content of all the news in Febrauary 2023

```python
## Let's extract all the links on the first page
response = requests.get('https://www.ghanaweb.com/GhanaHomePage/politics/browse.archive.php?date=20230226', headers=headers)
soup = bs(response.content)

div_class = soup.find('div', class_='bottom_artl_wrap')
print(div_class.prettify())
```

```python
li_tags = div_class.select('li a')
all_Links = []
for index, tag in enumerate(li_tags):
    all_Links.append(li_tags[index]['href'])
```

```python
all_Links
```

```python
dates = []
for i in reversed(range(10,27)):
    dates.append(str(i))
```

```python
dates
```

```python
final_dates = dates + ['09','08','07', '06', '05', '04', '03', '02', '01']
final_dates
```

```python
all_news_links = []

for date in final_dates:
    base_path = 'https://www.ghanaweb.com/GhanaHomePage/politics/browse.archive.php?date=202302'
    Full_Path = base_path + date
    url = Full_Path
    response = requests.get(url, headers=headers)
    webpage = bs(response.content)
    div_class = webpage.find('div', class_='bottom_artl_wrap')
    li_tags = div_class.select('li a')
    for index, tag in enumerate(li_tags):
        all_news_links.append(li_tags[index]['href'])
```

```python
## Extracting all the content of all the news links of each page

Content = []
Final_Content = []

for link in Full_Path:
    url = link
    response = requests.get(url, headers=headers)
    webpage = bs(response.content)
    Text = webpage.find_all('p', attrs={'id': 'article-123'})
    for index, text in enumerate(Text):
        Content.append(Text[index].get_text())
        
for index, content in enumerate(Content):
    Final_Content.append(content.replace('\n', '').replace('\r', ''))
```
